```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
hotel = "Resources/Datafiniti_Hotel_Reviews.csv"
hotel_df = pd.read_csv(hotel)
hotel_df.head()
```

```python
hotel2 = "Resources/Datafiniti_Hotel_Reviews_Jun19.csv"
hotel2_df = pd.read_csv(hotel2)
hotel2_df.head()
```

```python
hotel3 = "Resources/7282_1.csv"
hotel3_df = pd.read_csv(hotel3)
hotel3_df.head()
```

### Transform 7282_1  DataFrame

```python
# Create a filtered dataframe from specific columns for Hotel info
hot3 = ["name","address","city","postalCode","province","latitude","longitude","categories","reviews.rating","reviews.text"]
hotel3_transformed = hotel3_df[hot3].copy()

# Rename the Column Headers
hotel3_transformed = hotel3_transformed.rename(columns={"name":"Name","address":"Address","city":"City",
                                                        "postalCode":"Postal Code","province":"Province",
                                                        "latitude":"Latitude","longitude":"Longitude",
                                                        "categories":"Category","reviews.rating":"Review Rating",
                                                        "reviews.text":"Review Comments"})


hotel3_transformed.set_index("Name",inplace=True)

hotel3_transformed.head()
```

### Transform Datafiniti_Hotel_Reviews DataFrame

```python
# Create a filtered dataframe from specific columns for Pokemon
hot = ["name","address","city","province","postalCode","country","longitude","latitude","categories","reviews.rating","reviews.text","websites"]
hotel_transformed = hotel_df[hot].copy()

# Rename the Column Headers
hotel_transformed = hotel_transformed.rename(columns={"name":"Name","address":"Address","city":"City","province":"State",
                                                        "postalCode":"Postal Code","country":"Country","latitude":"Latitude",
                                                        "longitude":"Longitude","categories":"Category","reviews.rating":"Review Rating",
                                                        "reviews.text":"Review Comments"})


hotel_transformed.set_index("Name",inplace=True)

hotel_transformed.head()
```

### Transform Datafiniti_Hotel_Reviews_Jun19. DataFrame

```python
# Create a filtered dataframe from specific columns for Pokemon
hot2 = ["name","address","city","province","postalCode","country","longitude","latitude","categories","reviews.rating","reviews.text","websites"]
hotel2_transformed = hotel2_df[hot].copy()

# Rename the Column Headers
hotel2_transformed = hotel2_transformed.rename(columns={"name":"Name","address":"Address","city":"City","province":"State",
                                                        "postalCode":"Postal Code","country":"Country","latitude":"Latitude",
                                                        "longitude":"Longitude","categories":"Category","reviews.rating":"Review Rating",
                                                        "reviews.text":"Review Comments"})


hotel2_transformed.set_index("Name",inplace=True)

hotel2_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/hotels_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Create database connection

```python
hotel_transformed.to_sql(name='hotel', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from "hotel"', con=engine).head()
```

```python
hotel2_transformed.to_sql(name='hotel2', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from "hotel2"', con=engine).head()
```
