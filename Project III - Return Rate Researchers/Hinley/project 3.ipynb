```python
hotel_combo = pd.concat([hotel, hotel2]).drop_duplicates()
hotel_map=hotel_combo[['Name','Address','City','State','Postal Code','Longitude','Latitude']].dropna()
hotel_map['Latitude'] = hotel_map['Latitude'].astype(float)
hotel_map['Longitude'] = hotel_map['Longitude'].astype(float)
hotel_map
```

```python
def df_to_geojson(df, properties, lat='Latitude', lon='Longitude'):
    # create a new python dict to contain our geojson data, using geojson format
    geojson = {'type':'FeatureCollection', 'features':[]}

    # loop through each row in the dataframe and convert each row to geojson format
    
    for _, row in df.iterrows():
        # create a feature template to fill in
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}

        # fill in the coordinates
       # try:
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        #except KeyError:
         #   print(row)
            
        # for each column, get the value and add it as a new feature property
        for prop in properties:
            feature['properties'][prop] = row[prop]
        
        # add this feature (aka, converted dataframe row) to the list of features inside our dict
        geojson['features'].append(feature)
    
    return geojson
```

```python
cols = ['Name','Address','City','State','Postal Code']

geojson = df_to_geojson(hotel_map, cols)
```

```python
# save the geojson result to a file
output_filename = 'hotel.js'
with open(output_filename, 'w') as output_file:
    output_file.write('var dataset = {};'.format(json.dumps(geojson)))
    
# how many features did we save to the geojson file?
print('{} geotagged features saved to file'.format(len(geojson['features'])))
```
